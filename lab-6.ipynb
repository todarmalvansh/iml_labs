**1) to convert a NumPy array to a Pandas series**

```python
import numpy as np
import pandas as pd
import random
```

```python
array = np.array([11, 20, 1, 2,3, 4, 5, 6, 7])
  
print("Numpy array is :",array)

series = pd.Series(array) 
    
print("Pandas Series :" )
print(series)
```

**2) to convert the first column of a DataFrame as a Series**

```python
import numpy as np
import pandas as pd
data_frame = pd.DataFrame({'Enrollment': [37, 36, 46], 'SPI': [8.43,  8.94, 8.86]})
print("Data Frame : \n",data_frame,sep="");
series = pd.Series(data_frame.iloc[:,0])
print("\nSeries : \n",series,sep="")
```

**3) to create the mean and standard deviation of the data of a given Series**

```python
import numpy as np
import pandas as pd
import random
series2 = pd.Series(np.random.randint(1,20,5))
print("Pandas Series :")
print(series2)
print("Mean : ",series2.mean())
print("Standard Deviation : ",series2.std())
```

**4) to sort a given Series**

```python
import numpy as np
import pandas as pd
import random
series2 = pd.Series(np.random.randint(1,20,5))
print("Pandas Series :")
print(series2)
print("\nsorted Series : \n",series2.sort_values(ascending=True),sep="")
```
